#### 数据预处理

```python
import pandas as pd
from autogluon.tabular import TabularPredictor
```

```python
LABEL_HEADER = "SalePrice"
TRAIN_CSV_PATH = "../datasets/回归训练集.csv"
TEST_CSV_PATH = "../datasets/回归测试集.csv"
OUTPUT_MODEL_PATH = "../models/tabular_regression"  # 模型保存路径
```

```python
df_train = pd.read_csv(TRAIN_CSV_PATH)

df_train
```

```python
df_test = pd.read_csv(TEST_CSV_PATH)

df_test
```

```python
df_train.drop('Id', axis=1, inplace=True)
df_test.drop('Id', axis=1, inplace=True)
```

#### 模型拟合

```python
predictor = TabularPredictor(label=LABEL_HEADER, path=OUTPUT_MODEL_PATH).fit(
    df_train, presets="experimental", time_limit=120
)
```

```python
predictor.leaderboard()
```

#### 模型推理

```python
predictor = TabularPredictor.load(OUTPUT_MODEL_PATH)
y_pred = predictor.predict(df_test)
```
