#### 数据预处理

```python
import pandas as pd
from autogluon.tabular import TabularPredictor
```

```python
LABEL_HEADER = "Transported"
TRAIN_CSV_PATH = "../datasets/分类训练集.csv"
TEST_CSV_PATH = "../datasets/分类测试集.csv"
OUTPUT_MODEL_PATH = "../models/classification"  # 模型保存路径
```

```python
df_train = pd.read_csv(TRAIN_CSV_PATH)

df_train
```

```python
df_train[LABEL_HEADER] = df_train[LABEL_HEADER].apply(lambda x: 1 if x else 0)
```

```python
df_test = pd.read_csv(TEST_CSV_PATH)

df_test
```

```python
df_train.drop('PassengerId', axis=1, inplace=True)
df_test.drop('PassengerId', axis=1, inplace=True)
```

```python
df_train[['First_Name', 'Last_Name']] = df_train['Name'].str.split(' ', expand=True)
df_test[['First_Name', 'Last_Name']] = df_test['Name'].str.split(' ', expand=True)
```

```python
df_train[['Cabin_1', 'Cabin_2', 'Cabin_3']] = df_train['Cabin'].str.split('/', expand=True)
df_test[['Cabin_1', 'Cabin_2', 'Cabin_3']] = df_test['Cabin'].str.split('/', expand=True)
```

#### 模型训练

```python
# 启用最佳精度模式
predictor = TabularPredictor(label=LABEL_HEADER, path=OUTPUT_MODEL_PATH).fit(
    df_train, presets='experimental', time_limit=300
)
```

```python
predictor.leaderboard()
```

#### 模型推理

```python
predictor = TabularPredictor.load(OUTPUT_MODEL_PATH)
y_pred = predictor.predict(df_test)
```
